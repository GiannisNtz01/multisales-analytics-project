## Check whether the standard_cost column in the product table is the same as the cost column in the sales_product table.
To ensure that we can exclude the standard_cost column from the gold layer.

```python
sales = spark.table("sales_lakehouse.silver.sales")
```

```python
product = spark.table("sales_lakehouse.silver.product")
```

```python
sales_product = product.select("product_key", "standard_cost").join(sales.select("quantity" , "cost", "product_key"), on="product_key", how="left")
display(sales_product)
```

```python
from pyspark.sql.functions import col, abs

sales_product.filter(abs(((col("quantity") * col("standard_cost"))- (col("cost"))))>=1).count()
```

**We can see that only one product (439) shows a data inconsistency between the calculated cost (quantity * standard_cost) and the cost recorded in the sales table.**

```python

sales_product.filter(abs(((col("quantity") * col("standard_cost"))- (col("cost"))))>=1).show()
```
