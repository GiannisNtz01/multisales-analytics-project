### Product Keys in Sales that do not exist in Product

```python
%sql
SELECT 
    p.product_key,
    p.product
FROM sales_lakehouse.silver.product p
LEFT JOIN sales_lakehouse.silver.sales s
    ON p.product_key = s.product_key
WHERE s.product_key IS NULL;
```
