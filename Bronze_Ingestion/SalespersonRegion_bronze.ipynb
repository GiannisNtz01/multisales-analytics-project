## Read file

```python
df = spark.read.csv("/Volumes/lakehouse/adventureworks_multisales/raw_zone/SalespersonRegion.csv")
display(df)
```

```python
path = "/Volumes/lakehouse/adventureworks_multisales/raw_zone/SalespersonRegion.csv"

SalespersonRegion_uncleaned = (spark.read
      .format("csv")
      .option("header", "true")
      .option("sep", "\t")          # TAB delimiter
      .option("inferSchema", "true")
      .option("mode", "PERMISSIVE")
      .load(path))

display(SalespersonRegion_uncleaned)
```

```python
SalespersonRegion_uncleaned.printSchema()
```

## Show Nulls

```python
from pyspark.sql import functions as F

count_nulls = SalespersonRegion_uncleaned.select([F.sum(F.col(c).isNull().cast("int")).alias(c + "_nulls")
    for c in SalespersonRegion_uncleaned.columns])
display(count_nulls)
```

**show duplicates**

```python
total_rows = SalespersonRegion_uncleaned.count()
unique_rows = SalespersonRegion_uncleaned.dropDuplicates().count()

duplicate_count = total_rows - unique_rows
print("Total duplicated rows:", duplicate_count)
```

## whitespaces

```python
from pyspark.sql import functions as F

string_cols = [c for c, t in df.dtypes if t == "string"]

for c in string_cols:
    bad = df.filter(
        (F.col(c) != F.trim(F.col(c))) |   # leading/trailing
        (F.col(c).contains("  "))          # double spaces
    ).count()
    print(c, "bad_rows:", bad)
```
