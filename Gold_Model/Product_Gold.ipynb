### Gold layer. Making Product to dim_product. We will not insert the columns about colors

```python
product= spark.read.table("sales_lakehouse.silver.product")
```

```python
dim_product = product.select(
    "product_key", 
    "product", 
    "standard_cost",
    "color", 
    "subcategory",
    "category"
)

(dim_product.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true") #if we want to change any column later
    .saveAsTable("sales_lakehouse.gold.dim_product"))
```
