```python
region = spark.read.table("sales_lakehouse.silver.region")
```

```python
dim_region = region.select(
    "sales_territory_key", 
    "region", 
    "country", 
    "group"
)
(dim_region.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true") #if we want to change any column later
    .saveAsTable("sales_lakehouse.gold.dim_region"))
```
